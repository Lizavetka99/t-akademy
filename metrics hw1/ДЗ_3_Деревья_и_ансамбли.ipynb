```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/avia_price_data.csv")
data.head()
```

```python
X = data["avia_payments_rub"].astype(float)
Y = data["avia_payments_cnt"].astype(float)

R = X.sum() / Y.sum()
Ybar = Y.mean()

W = R + (X - R * Y) / Ybar
mu = W.mean()
sigma = W.std(ddof=1)

print('mu: ', mu)
print('sigma: ', sigma)
```
